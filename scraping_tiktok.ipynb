```python
import pandas as pd
pd.set_option('display.max_colwidth', None)
import os

from google_play_scraper import reviews, Sort
```

```python
# import files from gdrive
from google.colab import drive
drive.mount('/content/drive')
```

```python
file_path = '/content/drive/MyDrive/submission/NLP'
```

```python
ambil_data = reviews(
  'com.ss.android.ugc.trill', #id tiktok
  lang='id',
  country='id',
  sort=Sort.MOST_RELEVANT,
  count=20000
)
```

```python
tiktok_reviews = pd.DataFrame(ambil_data[0])
tiktok_reviews = tiktok_reviews[['content', 'score']]
tiktok_reviews
```

```python
f'{file_path}/data/tiktok_reviews.csv'
```

```python
os.makedirs(f'{file_path}/data', exist_ok=True)
tiktok_reviews.to_csv(f'{file_path}/data/tiktok_reviews.csv', index=False)
'Save ke csv selesai!!'
```
